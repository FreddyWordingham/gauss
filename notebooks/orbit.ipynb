```python
import math
import os
import numpy as np
import pandas as pd
import pymc3 as pm
cov = pm.gp.cov
import arviz as az
import matplotlib.pyplot as plt
import seaborn as sns

os.environ['MKL_NUM_THREADS'] = '1'
os.environ['OPENBLAS_NUM_THREADS'] = '1'

sns.set_context("notebook")
```

```python
# data = pd.read_csv('../data/orbit.csv', index_col=0);
data = pd.read_table('../data/orbit.csv', sep=',')
data.plot.scatter(x='mjd', y='rv');
x, y = data[['mjd', 'rv']].values.T;

def f(x):
    return math.sin(x / 250.0)

def array_map(x):
    return np.array(list(map(f, x)))

x = array_map(x);
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x, y, data=data, alpha=0.3);
```
